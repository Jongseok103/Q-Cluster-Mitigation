```python
import pandas as pd
import json
import torch
from tqdm import tqdm
from bert_score import score as bert_score
from comet import download_model, load_from_checkpoint
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import matplotlib.pyplot as plt
import seaborn as sns

# ==============================================================================
# 1. 설정 및 데이터 로드
# ==============================================================================
INPUT_FILE = "generation_results.json"  # 이전 단계에서 저장한 파일
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

print(f"📂 데이터 로드 중: {INPUT_FILE}")
with open(INPUT_FILE, 'r', encoding='utf-8') as f:
    data = json.load(f)

df = pd.DataFrame(data)
print(f"📊 데이터 크기: {len(df)} rows")

# ==============================================================================
# 2. Metric 1: BERTScore (Reference vs Candidate) - 한국어
# ==============================================================================
print("\n🚀 [1/3] BERTScore 계산 중 (Korean)...")
# bert-base-multilingual-cased 모델을 사용하여 다국어(한국어 포함) 유사도 측정
# hash code 리턴은 무시하기 위해 _ 사용
P, R, F1 = bert_score(
    df['candidate'].tolist(),
    df['reference'].tolist(),
    lang="ko",
    verbose=True,
    device=DEVICE,
    model_type='bert-base-multilingual-cased'
)

df['bert_score'] = F1.tolist()
print("✅ BERTScore 계산 완료")

# ==============================================================================
# 3. Metric 2: COMET (Source + Reference + Candidate)
# ==============================================================================
print("\n🚀 [2/3] COMET 점수 계산 중 (wmt22-comet-da)...")
# COMET 모델 다운로드 및 로드 (처음 실행 시 다운로드 시간 소요)
try:
    model_path = download_model("Unbabel/wmt22-comet-da")
    comet_model = load_from_checkpoint(model_path)

    # COMET 입력 형식에 맞게 데이터 변환
    comet_data = [
        {"src": row['source'], "mt": row['candidate'], "ref": row['reference']}
        for _, row in df.iterrows()
    ]

    # 배치 단위로 추론
    comet_output = comet_model.predict(comet_data, batch_size=8, gpus=1 if "cuda" in DEVICE else 0)
    df['comet_score'] = comet_output.scores
    print("✅ COMET 점수 계산 완료")

except Exception as e:
    print(f"⚠️ COMET 실행 실패 (설치/환경 문제 가능성): {e}")
    df['comet_score'] = 0.0

# ==============================================================================
# 4. Metric 3: 역번역 유사도 (Back Translation Similarity)
# ==============================================================================
print("\n🚀 [3/3] 역번역 유사도 측정 중 (Ko -> En)...")

# 역번역을 위한 별도 모델 로드 (Helsinki-NLP/opus-mt-ko-en : 가볍고 성능 준수)
back_trans_model_name = "Helsinki-NLP/opus-mt-ko-en"
tokenizer = AutoTokenizer.from_pretrained(back_trans_model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(back_trans_model_name).to(DEVICE)

def back_translate(text_list):
    inputs = tokenizer(text_list, return_tensors="pt", padding=True, truncation=True, max_length=128).to(DEVICE)
    with torch.no_grad():
        outputs = model.generate(**inputs, max_length=128)
    return tokenizer.batch_decode(outputs, skip_special_tokens=True)

# 배치 처리로 역번역 수행 (속도 최적화)
batch_size = 16
back_translated_texts = []

for i in tqdm(range(0, len(df), batch_size), desc="Back-Translating"):
    batch_texts = df['candidate'].iloc[i : i + batch_size].tolist()
    back_translated_texts.extend(back_translate(batch_texts))

df['back_translated_en'] = back_translated_texts

# 역번역된 영어(En_back)와 원문 영어(En_original) 간의 BERTScore 측정 (English)
print("   -> 역번역 문장과 원문 간의 BERTScore(En) 측정 중...")
P_back, R_back, F1_back = bert_score(
    df['back_translated_en'].tolist(),
    df['source'].tolist(),
    lang="en",
    verbose=True,
    device=DEVICE,
    model_type='roberta-large' # 영어는 roberta-large가 성능이 좋음
)

df['back_trans_similarity'] = F1_back.tolist()
print("✅ 역번역 유사도 계산 완료")

# ==============================================================================
# 5. 결과 저장 및 시각화
# ==============================================================================
# CSV 저장
output_csv = "quantitative_evaluation_results.csv"
df.to_csv(output_csv, index=False, encoding='utf-8-sig')
print(f"\n💾 전체 결과 저장 완료: {output_csv}")

# 모델별 평균 계산
summary = df.groupby('model_type')[['bert_score', 'comet_score', 'back_trans_similarity']].mean()
print("\n📊 [모델별 정량 평가 평균]")
print(summary)

# 시각화 (Bar Chart)
plt.figure(figsize=(12, 6))
summary_melted = summary.reset_index().melt(id_vars='model_type', var_name='Metric', value_name='Score')

sns.barplot(data=summary_melted, x='Metric', y='Score', hue='model_type', palette='viridis')
plt.title("Quantitative Metrics Comparison (Base vs Tuned)")
plt.ylim(0, 1.1) # 대부분의 점수는 0~1 사이 (COMET은 다를 수 있음)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend(title='Model Type')
plt.show()
```
